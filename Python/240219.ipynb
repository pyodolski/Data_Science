#### 멜론 차트 종합

```python
import time 
from selenium import webdriver
driver = webdriver.Chrome()

url = 'https://www.melon.com/chart/index.htm'
driver.get(url)

time.sleep(2)

search_query = 'Dynamite'
search_box = driver.find_elements('css selector', '#top_search')
# len(search_box)
search_box[0].clear()
search_box[0].send_keys(search_query)
search_box[0].send_keys(Keys.RETURN)

time.sleep(2)

from bs4 import BeautifulSoup 
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

titles = soup.select('a.fc_gray')
singers = soup.select('div#artistName > a.fc_mgray')

for title, singer in zip(titles, singers):
    print(title.text, singer.text , sep=' | ')
```

```python
# 차트이름, 순위, 노래제목, 가수
# Melon, 1위, Love, 악뮤

import time 
from selenium import webdriver
from bs4 import BeautifulSoup 
driver = webdriver.Chrome()

url = 'https://www.melon.com/chart/index.htm'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

```

```python
songs= soup.select('tbody > tr')
len(songs)
```

```python
song = songs[0]
rank = song.select('div > .rank') # 랭크 정보 가져오기
```

```python
song_data =[]

for song in songs:
    rank_ = f'{song.select("div > .rank")[0].text}위'
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    song_data.append(['Melon', rank_, title, singer])

song_data
```

```python
import pandas as pd
columns = ['서비스', '순위', '타이틀', '가수']
pd_data = pd.DataFrame(song_data)
pd_data.head()
```

```python
pd_data.to_excel('../ch_02/data/melon.xlsx', index = False)
```
