```python
total_tags = []
for tag_list in raw_total['tags'].values:
  tag_ = tag_list[1:-1].split(',')
  for tag in tag_:
    # "'제주여행'"
    total_tags.append(tag.replace("'","").strip())
    
total_tags
```

```python
len(total_tags)
```

```python
from collections import Counter # 문자가 몇번 나왔는지 탐색해주는 import
tag_count = Counter(total_tags)
print(tag_count.most_common(30))
```

```python
# 의미없는 Tag들을 제거하자
black_tag = ['제미녀', '제미녀_서귀포시', '제주도에미친녀자', '제주그램', '제주도그램', '']

final_tag = []
for tag in total_tags:
  if tag not in black_tag:
    final_tag.append(tag)
```

```python
tag_count = Counter(final_tag)
tag_count.most_common(10)
```

```python
len(tag_count)
```

```python
type(tag_count)
tag_count
```

```python
df_ = pd.DataFrame(tag_count.most_common(30))
df_.columns = ["tag", "counts"]
df_.head(), df_.info()
```

```python
df_.head()
```

```python
df_ = pd.DataFrame({'tags':final_tag})
df_ = df_['tags'].value_counts().reset_index()
df_.columns = ['tag', 'count']
df_.head()

df_.info()
```

```python
tag_top_50 = df_.head(50)
tag_top_50.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic' #Malgun Gothic
plt.figure(figsize=(14, 10))
sns.barplot(x='count', y='tag', data=tag_top_50, palette='Set2')
plt.show()
```
