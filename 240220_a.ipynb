## 벅스 차트 연결 및 html 파싱

```python
driver = webdriver.Chrome() # 드라이버 크롬으로 설정
```

```python
url = 'https://music.bugs.co.kr/chart' # 벅스차트로 연결 
driver.get(url)
```

```python
html = driver.page_source # 활성화 되어있는 페이지의 html소스를 변수에 저장
soup = BeautifulSoup(html, 'html.parser')
# eautifulSoup 라이브러리를 사용하여 HTML 문서를 파싱하여 파이썬에서 다루기 쉬운 객체로 만드는 작업
```

### 탐색

```python
songs = soup.select('table.byChart > tbody > tr') 
# table 태그 안의 byChart 클래스 안의 tbody 안의 tr 개수 탐색
len(songs)
```

```python
# 밤양갱 노래 제목 출력하기
song = songs[0] # 첫번째 태그를 가져와서 변수에 저장
title = song.select('p.title > a')[0].text # p 태그의 title 클래스 안의 a 태그 불러오기
title # 출력하기 전 길이 출력해서 원하는 값 '1개' 인지 확인하고 출력
```

```python
# 순위 출력하기
rank = song.select('div.ranking > strong')[0].text
rank = f'{rank}위'
rank
```

```python
# 가수이름 출력하기
singer = song.select('p.artist > a')[0].text
singer
```

#### 차트 만들기

```python
song_data = [] 

songs = soup.select('table.byChart > tbody > tr') 
for song in songs:
    rank = song.select('div.ranking > strong')[0].text
    rank = f'{rank}위'
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text

    song_data.append(['Bugs', rank, title, singer])

song_data[:10]

    # 각 노래마다 노래제목, 가수이름, 순위가 있기 때문에 반복문 = song
```

#### pandas 로 차트 만들기

```python
columns = ['서비스', '순위', '타이틀', '가수'] 
pd_ = pd.DataFrame(song_data, columns=columns) # 데이터 삽입
# 2차원 배열 생성
pd_ 
```

##### 엑셀파일 저장

```python
pd_.to_excel('../data/bugs.xlsx', index = False) # 불필요한 인덱스 삭제
```

## 지니 차트 연결 및 html 파싱

```python
url = 'https://www.genie.co.kr/chart/top200' # 지니차트로 연결 
driver.get(url)
```

```python
songs = soup.select('table.list-wrap > tbody > tr') 
songs
```

```python
song = songs[0] 
title = song.select('td.info > a.title')[0].text.strip()
title 
```

```python
singer = song.select('td.info > a.artist')[0].text.strip()
singer
```

```python
song_data = [] 

songs = soup.select('table.list-wrap > tbody > tr') 
for song in songs:
    rank = song.select('tr.list > td.number')[0].text
    rank = rank.split('\n')[0] 
    rank = f'{rank}위'
    title = song.select('td.info > a.title')[0].text.strip()
    singer = song.select('td.info > a.artist')[0].text

    song_data.append(['Genie', rank, title, singer])

song_data[:50]
```

### 사이트 변경

```python
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240220&hh=10&rtm=Y&pg=2
song_data = [] 

driver = webdriver.Chrome()
i = 1
while (i <= 4):
    url = f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20240220&hh=10&rtm=Y&pg={i}'
    i += 1
    driver.get(url)
    time.sleep(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

   

    songs = soup.select('tr.list') 
    
    for song in songs:
        rank = song.select('tr.list > td.number')[0].text
        rank = rank.split('\n')[0] 
        rank = f'{rank}위'
        title = song.select('td.info > a.title')[0].text.strip()
        singer = song.select('td.info > a.artist')[0].text

        song_data.append(['Genie', rank, title, singer])

song_data[:]
```

#### pandas 데이터 프레임 생성

```python
columns = ['서비스', '순위', '타이틀', '가수'] 
pd_ = pd.DataFrame(song_data, columns=columns) # 데이터 삽입
# 2차원 배열 생성
pd_ .head()
```
